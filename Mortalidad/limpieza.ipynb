```python
import pandas as pd

# Leer el archivo Excel
df = pd.read_excel("mortalidad2020.xlsx", engine="openpyxl")

# Copiar el DataFrame original
df_clean = df.copy()

# Renombrar la columna Unnamed: 0 a "Sexo"
df_clean = df_clean.rename(columns={"Unnamed: 0": "Sexo"})

# Reemplazar comas y convertir a números
for col in df_clean.columns[1:]:  # Ignorar la columna 'Sexo'
    df_clean[col] = df_clean[col].str.replace(',', '', regex=False)
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

# Guardar como nuevo archivo Excel
df_clean.to_excel("mortalidad2020_limpio.xlsx", index=False)
print("Archivo guardado como 'mortalidad2020_limpio.xlsx'")


```

```python
import pandas as pd

# 1. Cargar los archivos Excel
df_2020 = pd.read_excel("mortalidad2020_limpio.xlsx")
df_2021 = pd.read_excel("mortalidad2021_limpio.xlsx")
df_2022 = pd.read_excel("mortalidad2022_limpio.xlsx")
df_2023 = pd.read_excel("mortalidad2023_limpio.xlsx")

# 2. Agregar la columna "Año"
df_2020["Año"] = 2020
df_2021["Año"] = 2021
df_2022["Año"] = 2022
df_2023["Año"] = 2023

# 3. Unir los DataFrames
df_total = pd.concat([df_2020, df_2021, df_2022, df_2023], ignore_index=True)

# 4. Eliminar columna "Total" si existe
if "Total" in df_total.columns:
    df_total = df_total.drop(columns=["Total"])

# 5. Diccionario de claves por entidad
claves_entidades = {
    'Aguascalientes': '01',
    'Baja California': '02',
    'Baja California Sur': '03',
    'Campeche': '04',
    'Coahuila de Zaragoza': '05',
    'Colima': '06',
    'Chiapas': '07',
    'Chihuahua': '08',
    'Ciudad de México': '09',
    'Durango': '10',
    'Guanajuato': '11',
    'Guerrero': '12',
    'Hidalgo': '13',
    'Jalisco': '14',
    'México': '15',
    'Michoacán de Ocampo': '16',
    'Morelos': '17',
    'Nayarit': '18',
    'Nuevo León': '19',
    'Oaxaca': '20',
    'Puebla': '21',
    'Querétaro': '22',
    'Quintana Roo': '23',
    'San Luis Potosí': '24',
    'Sinaloa': '25',
    'Sonora': '26',
    'Tabasco': '27',
    'Tamaulipas': '28',
    'Tlaxcala': '29',
    'Veracruz de Ignacio de la Llave': '30',
    'Yucatán': '31',
    'Zacatecas': '32'
}

# 6. Renombrar columnas con claves
df_total = df_total.rename(columns=claves_entidades)

# 7. Reestructurar el DataFrame (melt + pivot)
df_melt = df_total.melt(id_vars=["Sexo", "Año"], var_name="CLAVE", value_name="MUERTES")
df_pivot = df_melt.pivot_table(index=["Año", "CLAVE"], columns="Sexo", values="MUERTES", aggfunc="sum").reset_index()

# 8. Renombrar columnas finales
df_pivot.columns.name = None
df_pivot = df_pivot.rename(columns={
    "Hombre": "HOMBRE",
    "Mujer": "MUJER",
    "No especificado": "SIN ESPECIFICAR"
})

# 9. Reordenar columnas
df_final = df_pivot[["Año", "CLAVE", "MUJER", "HOMBRE", "SIN ESPECIFICAR"]]

# 10. Guardar resultado
df_final.to_excel("mortalidad_total_2020_2023_estructurado.xlsx", index=False)

print("Archivo generado: mortalidad_total_2020_2023_estructurado.xlsx")
```
